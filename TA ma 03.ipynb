```python
def CreateCorpusFromDataFrame(corpusfolder,df): # The new corpus's path and the data frame it is created from. 
    """
    This function takes a data frame containing text and transforms it into a corpus.
    Each row of the excel becomes a document. 
    The name of the document takes the fileid and the class it belongs to. 
    """
    for index, r in df.iterrows():
        doc_content=r['doc_content']
        doc_id=r['doc_id']
        
        fname=str(doc_id)+'.txt' # Document name practically
        corpusfile=open(corpusfolder+'/'+ fname,'w',encoding="utf-8") # a for appending; creates file if non-existent
        corpusfile.write(str(doc_content)) # write the document
        corpusfile.close() # and close it 


CreateCorpusFromDataFrame(r"C:\Users\aleen\Desktop\Data Science\Text Analytics\Assignments\Corpora\Wine Reviews",df)
```

#### Pre-processing phase

```python
def tokenizer(text):
    text = re.sub("[^a-zA-Z]+", " ", text)
    tokens = nltk.tokenize.word_tokenize(text)
    return tokens  
```

```python
def preprocessing_txt(text):
    
    tokens = tokenizer(text)
    stemmer = nltk.stem.porter.PorterStemmer()
    stopwords = nltk.corpus.stopwords.words('english')
    new_text = ""
    for token in tokens:
        token = token.lower()
        if token not in stopwords:
#             print token
            new_text += stemmer.stem(token)
            new_text += " "
        
    return new_text
```

### Query function

```python
def process_query(query_text, dictionary):
    
    result = []
    query = preprocessing_txt(query_text)
    query = query.split()
    #print(query)
    for i,word in enumerate(query):
        #print(i)
        if word not in dictionary.keys():
            return ("No such word(s) found.")
        else: 
            if i == 0:
                for i in dictionary[word].items():
                    result.append(i[0])
            else:
                temp = []
                for i in dictionary[word].items():
                    temp.append(i[0])
                result = list(set(result).intersection(set(temp)))
            for res in result:
                str(res) + "\n"
    return result
```
